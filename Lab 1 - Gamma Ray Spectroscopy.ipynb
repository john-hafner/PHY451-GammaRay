$\gamma$-rays interact with matter in a few ways per the lab notes, the photoelectric effect, compton scattering, and pair production.

## Compton Scattering
Starting with compton scattering, the $\gamma$-ray will impact an electron or other particle, and impart some of its kinetic energy to the particle.  As a result it'll have lower energy itself, and we can detect that $\Delta\lambda$.
$$\Delta\lambda = \lambda^{'} - \lambda = \lambda_c(1 - cos(\theta))$$<br>
$$\lambda_c = \frac{h}{m_ec}$$

We won't know the $\theta$ but the kinetic energy should be largest with a $180^{\circ}$ scattering angle.  We can see that with some graphing.
