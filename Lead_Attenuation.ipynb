We did the lead attenuation by putting progressively larger squares of lead on top of the source, and then measured the activity using MCS mode on the MCA software.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
lead_v2_url = 'https://raw.githubusercontent.com/john-hafner/PHY451-GammaRay/main/data/lead_v2.txt'
leadv2 = 'https://raw.githubusercontent.com/john-hafner/PHY451-GammaRay/main/data/lead_v2.txt'
lead_v1_url = 'https://raw.githubusercontent.com/john-hafner/PHY451-GammaRay/main/data/lead_v1.txt'
df = pd.read_csv(lead_v1_url)
print(f'Size of data: {df.shape}')
df.head()
```

```python
plt.figure(figsize=(10,8))

plt.plot(df)
plt.title('Lead Activity (v1)')
plt.ylabel('Counts')
plt.xlabel('Time')
plt.show()
```

```python
x = df['count'].to_numpy()
delta = abs(x[5:] - x[:-5])

plt.figure(figsize=(20,8))
plt.plot(np.arange(0, len(delta)), delta)
plt.title('$\Delta$(count)')
plt.ylabel('$\Delta$(count)')
plt.xlabel('time')
```
